```python
# bfs 

from itertools import combinations
from collections import deque
import copy

n, m = map(int, input().split())
graph = []
for _ in range(n):
    graph.append(list(map(int, input().split())))
    
none = []
virus = []
for i in range(n):
    for j in range(m):
        if graph[i][j] == 0:
            none.append([i,j])
        elif graph[i][j] == 2:
            virus.append([i,j])
        
            
# none에서 3개 조합하기
candidates = list(combinations(none, 3))

def safe_area(graph_new):
    safe = 0
    for i in range(n):
        for j in range(m):
            if graph_new[i][j] == 0:
                safe += 1
    return safe

dx = [-1,1,0,0]
dy = [0,0,-1,1]

safe_area_max = 0
for candidate in candidates:
    graph_new = copy.deepcopy(graph)
    for x, y in candidate:
        graph_new[x][y] = 1
        
    # bfs 시작
    for v in virus:
        q = deque()
        q.append(v)

        while q:
            now_x, now_y = q.popleft()
            for i in range(4):
                nx = now_x + dx[i]
                ny = now_y + dy[i]
                if 0<=nx<n and 0<=ny<m and graph_new[nx][ny] == 0:
                    graph_new[nx][ny] = 2
                    q.append([nx, ny])

    safe_area_max = max(safe_area_max, safe_area(graph_new))
    
print(safe_area_max)
```
