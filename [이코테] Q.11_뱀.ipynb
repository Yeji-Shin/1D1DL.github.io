```python
# 풀이 1

n = int(input())  # 보드의 크기 n*n
k = int(input())  # 사과의 개수 k
graph = [[0] * (n+1) for _ in range(n+1)]  # 그래프 정보
for _ in range(k):
    x, y = map(int, input().split())
    graph[x][y] = 1  # 사과의 위치 1로 표시 

l = int(input()) # 뱀의 방향 변환 횟수
change_dir = {}
for _ in range(l):
    time, dir = input().split()
    change_dir[int(time)] = dir
    
from collections import deque

snake = deque((1,1))
# 처음 뱀 위치
x = 1
y = 1

dir_now = 0   # 오른쪽 방향
# 우하좌상 순서
dx = [0, 1, 0, -1]
dy = [1, 0, -1, 0]

t = 0 

while True:
            
    nx = x + dx[dir_now]
    ny = y + dy[dir_now]
    
    # 범위를 벗어나거나 몸이 부딫히면 stop 
    if nx <1 or nx >n or ny<1 or ny>n or (nx, ny) in snake: 
        t += 1
        print(t)
        break
    else:
        snake.append((nx,ny))
        if graph[nx][ny] == 1:  # 사과가 있으면 꼬리 움직이지 않기 
            graph[nx][ny] = 0
        else:
            snake.popleft()  # 사과가 없으면 꼬리를 줄이기
    
    x, y = nx, ny  # 다음 위치로 머리 이동
    t += 1
    
    if t in list(change_dir.keys()):
        if change_dir[t] == 'D':
            dir_now += 1
        else:
            dir_now -= 1
    
```

```python
# 풀이 2

n = int(input())  # 보드의 크기 n*n
k = int(input())  # 사과의 개수 k

graph = [[0] * (n+1) for _ in range(n+1)]  # 그래프 정보
for _ in range(k):
    x, y = map(int, input().split())
    graph[x][y] = 1  # 사과의 위치 1로 표시 

l = int(input()) # 뱀의 방향 변환 횟수
change_dir = []
for _ in range(l):
    time, dir = input().split()
    change_dir.append((int(time), dir))
    
dir_now = 0   # 처음에 오른쪽 방향
# 우하좌상 순서
dx = [0, 1, 0, -1]
dy = [1, 0, -1, 0]

def turn(direction, c):
    if c == 'L':
        direction = (direction-1) % 4
    else:
        direction = (direction+1) % 4
    return direction

def simulate():
    x, y = 1, 1  # 뱀의 초기 위치
    graph[x][y] = 2  # 뱀이 존재하는 위치 2로 표ㅕ시
    t = 0
    index = 0  # 다음에 회전할 정보
    dir_now = 0   # 머리의 방향
    snake = [(x,y)]   # 뱀이 존재하는 위치
    
    while True:
        nx = x + dx[dir_now]
        ny = y + dy[dir_now]
        
        if 1<=nx<=n and 1<=ny<=n and graph[nx][ny] != 2:
            if graph[nx][ny] == 1:  # 사과가 있으면 꼬리 움직이지 않기 
                graph[nx][ny] = 2
                snake.append((nx, ny))
            if graph[nx][ny] == 0:  # 사과가 없으면 꼬리를 줄이기
                graph[nx][ny] = 2
                snake.append((nx, ny))
                px, py = snake.pop(0)
                graph[px][py] = 0
        else:
            t += 1
            break
            
        x, y = nx, ny  # 다음 위치로 머리 이동
        t += 1
    
        if index < 1 and t == change_dir[index][0]:
            dir_now = turn(dir_now, change_dir[index][1])
            index += 1
    return time

print(simulate())
```
