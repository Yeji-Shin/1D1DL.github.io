```python
# 간선의 비용이 모두 1이기 때문에 bfs 문제

# 로봇이 차지하고 있는 두 칸은 튜플로 처리 (집합 자료형)
'''
이동: 상하좌우로 이동가능
회전
 - 로봇이 가로인 상태에서 아래쪽으로 회전: 아래쪽 두 칸모두 벽이 없으면 가능
 - 로봇이 가로인 상태에서 위쪽으로 회전: 위쪽 두 칸 모두 벽이 없으면 가능 
 - 로봇이 세로인 상태에서 왼쪽으로 회전: 왼쪽 두 칸 모두 벽이 없으면 가능
 - 로봇이 세로인 상태에서 오른쪽으로 외전: 오른쪽 두 칸 모두 벽이 없으면 가능 
'''

from collections import deque

# 다음번에 이동 가능한 위치들 알아보기
def get_next_pos(now_pos, board):
    next_pos = []   # 현재 위치에서 이동 가능한 위치들 
    now_pos = list(now_pos)  # 집합을 리스트로 변환
    x1, y1, x2, y2 = now_pos[0][0], now_pos[0][1], now_pos[1][0], now_pos[1][1]
    
    # 상하좌우로 이동하는 경우
    dx = [-1,1,0,0]
    dy = [0,0,-1,1]
    for i in range(4):
        nx1, ny1, nx2, ny2 = x1+dx[i], y1+dy[i], x2+dx[i], y2+dy[i]
        if board[nx1][ny1]==0 and board[nx2][ny2]==0:
            next_pos.append({(nx1,ny1), (nx2, ny2)})
            
    # 회전하는 경우
    # 로봇이 가로인 상태
    if x1 == x2:
        for i in [-1,1]:  # 위쪽으로 회전, 아래쪽으로 회전
            if board[x1+i][y1]==0 and board[x2+i][y2]==0:
                next_pos.append({(x1,y1), (x1+i,y1)})  # 왼쪽을 기준으로 위아래 회전
                next_pos.append({(x2,y2), (x2+i,y2)})  # 오른쪽을 기준으로 위아래 회전                
    # 로봇이 세로인 상태
    elif y1 == y2:
        for i in [-1,1]:  # 왼쪽으로 회전, 오른쪽으로 회전
            if board[x1][y1+i]==0 and board[x2][y2+i]==0:
                next_pos.append({(x1,y1), (x1,y1+i)})  # 위에 것을 기준으로 좌우 회전
                next_pos.append({(x2,y2), (x2,y2+i)})  # 아래 것을 기준으로 좌우 회전
                
    return next_pos


# bfs
def solution(board):
    # 쉽게 하기 위해 맵의 외곽에 벽을 두는 형태로 맵 변형
    n = len(board)
    new_board = [[1]*(n+2) for _ in range(n+2)]
    for i in range(n):
        for j in range(n):
            new_board[i+1][j+1] = board[i][j]
    
    q = deque()
    visited = []
    pos = {(1,1),(1,2)}
    q.append((pos,0))  # (현재 위치, cost)
    visited.append(pos)  # 방문처리
    
    while q:
        now_pos, cost = q.popleft()
        
        if (n,n) in now_pos:  # (n-1, n-1)위치에 도달했으면 최단 거리 반환
            return cost
        
        for next_pos in get_next_pos(now_pos, new_board):
            if next_pos not in visited:
                q.append((next_pos, cost+1))
                visited.append(next_pos)
                
    return 0
```
