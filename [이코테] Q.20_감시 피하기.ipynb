```python
# 풀이 2: DFS활용 

from itertools import combinations

n = int(input())
data = []
teachers = [] 
nones = []
for i in range(n):
    data.append(list(input().split()))
    for j in range(n):
        if data[i][j] == 'T':
            teachers.append((i,j))
        elif data[i][j] == 'X':
            nones.append((i,j))
            
obstacles = list(combinations(nones, 3))

def find(x,y):
    dx = [-1,1,0,0]
    dy = [0,0,-1,1]
    
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        if 0<=nx<n and 0<=ny<n:
            if data_new[nx][ny] == 'S':
                return True
                break
            elif data_new[nx][ny] == 'X':
                data_new[nx][ny] = 'T'
                find(nx,ny)
    return False
            
    
# 장애물 설치한 후 한 명이라도 학생이 감지되는지 검사
def process():
    for tx, ty in teachers:
        if find(tx,ty):
            return True
    return False

from copy import deepcopy
# 3개에 장애물 설치
answer = False
for obs in obstacles:
    for ox, oy in obs:
        data_new = deepcopy(data)
        data_new[ox][oy] = 'O'
        if not process():
            answer = True
            break

if answer:
    print('YES')
else:
    print('NO')
```
