```python
# 현재 위치에서 모든 물고기까지의 최단 거리를 구한 후 먹을 물고기의 위치를 찾기 
# 거리가 1인 최단 거리 알고리즘은 bfs 사용

n = int(input())  # 공간의 크기
graph = []  # 현재 그래프
for _ in range(n):
    graph.append(list(map(int, input().split())))
    
for i in range(n):
    for j in range(n):
        if graph[i][j] == 9:
            shark_x, shark_y = i, j
            graph[i][j] = 0  # 아기 상어의 현재 위치를 0으로 만들어주기 
            
shark_len = 2  # 초기의 상어 크기


# 최단 거리 알고리즘 bfs
from collections import deque

dx = [-1,1,0,0]
dy = [0,0,-1,1]

def bfs():
    q = deque()
    q.append((shark_x, shark_y))

    dist = [[-1] * (n) for _ in range(n)]  # 최단 거리 테이블
    dist[shark_x][shark_y] = 0
    
    while q:
        x, y = q.popleft()

        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if 0<=nx<n and 0<=ny<n: 
                if graph[nx][ny]<=shark_len and dist[nx][ny] == -1: # 자신의 크기보다 큰 물고기가 있는 칸은 지나갈 수 없음
                    dist[nx][ny] = dist[x][y] + 1
                    q.append((nx,ny))
    return dist


# 최단 거리 테이블을 가지고 먹을 물고기를 찾는 함수
def find(dist):
    x, y = 0, 0  # 먹을 물고기의 위치 초기화
    min_dist = int(1e9)  # 먹을 물고기까지의 거리 초기화
    
    # 거리가 같으면 가장 위에 있는 물고기 -> 가장 왼쪽에 있는 물고기
    for i in range(n):
        for j in range(n):
            if dist[i][j] != -1 and 1<=graph[i][j]<shark_len:   # 자신의 크기보다 작은 물고기만 먹을 수 있음
                if dist[i][j] < min_dist:  # 가장 가까운 물고기 1마리만 선택
                    x, y = i, j
                    min_dist = dist[i][j]
                    
    if min_dist == int(1e9):  # 먹을 수 있는 물고기가 없는 경우
        return None
    else:
        return x, y, min_dist
    
    
# 먹기 시작
eat = 0
result = 0

while True:
    fish_loc = find(bfs())  # 먹을 수 있는 물고기 위치
    
    if fish_loc == None:  # 먹을 수 있는 물고기가 더 이상 없을 때
        print(result)
        break
    else:
        shark_x, shark_y = fish_loc[0], fish_loc[1]  # 현재 위치 갱신
        result += fish_loc[2]
        
        graph[shark_x][shark_y] = 0  # 먹은 위치에는 아무 것도 없도록 처리
        
        eat += 1
        if eat >= shark_len:
            eat = 0
            shark_len += 1
```
