```python
# 위상 정렬: 우선순위에 따라 순서를 나열

'''
팀 간 순위 정보를 그래프로 표현한 뒤에 위상 정렬 알고리즘 사용

1. 자기보다 낮은 등수의 팀을 가리키도록 방향 그래프 설정 
2. 상대적 순위가 바뀌면 간선의 방향을 바꾸기
3. 새롭게 위상 정렬을 수행 
4. 사이클이 발생하지 않고 위상 정렬의 결과가 하나 일때만 possible 출력

위상 정렬의 결과가 하나 = 큐에서 노드를 뽑을 때 큐의 원소가 항상 1개로 유지되는 경우
사이클 발생은 서로소 집합으로 확인
'''

from collections import deque

for tc in range(int(input())):
    
    n = int(input())  # 팀의 수
    graph =[[False]*(n+1) for _ in range(n+1)]
    data = list(map(int, input().split()))
    indegree = [0] *  (n+1)
    for i in range(n):
        for j in range(i+1,n):
            graph[data[i]][data[j]] = True
            indegree[data[j]] += 1
            
    m = int(input())  # 등수가 바뀐 팀의 수 
    for i in range(m):
        a, b = map(int, input().split())
        # 간선 방향 뒤집기
        if graph[a][b]:
            graph[a][b] = False
            graph[b][a] = True
            indegree[a] += 1
            indegree[b] -= 1
        else:
            graph[a][b] = True
            graph[a][b] = False
            indegree[a] -= 1
            indegree[b] += 1
            
    # 위상 정렬
    result = []
    q = deque()
    
    for i in range(1, n+1):
        if indegree[i] == 0:
            q.append(i)
            
    only_one = True
    cycle = False
    
    for i in range(n):
        if len(q) == 0:
            cycle = True
            break
        if len(q) >= 2:
            only_one = False
            break
    
        now = q.popleft()
        result.append(now)
        for j in range(1,n+1):
            if graph[now][j]:
                indegree[j] -= 1
                if indegree[j] == 0:
                    q.append(j)
                    
    if cycle:
        print('IMPOSSIBLE')
    elif not only_one:
        print('?')
    else:
        for i in result:
            print(i, end= ' ')
        print()
```
