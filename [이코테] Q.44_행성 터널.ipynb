```python
# 풀이 1

# 최소 신장 트리

'''
비용이 주어지지 않았으므로 각 연결에 대해 모든 비용을 다 계산한 후
비용이 낮은 것 부터 연결하기 (사이클이 없는 경우에만)
'''

def find_parent(parent, x):
    if parent[x] != x:
        parent[x] = find_parent(parent, parent[x])
    return parent[x]

def union(parent, a, b):
    a = find_parent(parent, a)
    b = find_parent(parent, b)
    if a < b:
        parent[b] = a
    else:
        parent[a] = b

def calculate_cost(a, b):
    xa, ya, za = a
    xb, yb, zb = b
    return min(abs(xa-ya), abs(ya-yb), abs(za-zb))

n = int(input())  # 행성의 개수

planets = []
for _ in range(n):
    x, y, z = map(int, input().split())
    planets.append((x,y,z))
    
# 각 행성간의 거리 계산하기
edges = []
for i in range(n-1):
    for j in range(i+1, n):
        edges.append((calculate_cost(planets[i], planets[j]), i, j))
edges.sort()

parent = [0] * n
for i in range(n):
    parent[i] = i

result = 0
for edge in edges:
    cost, planetA, planetB = edge
    if find_parent(parent, planetA) != find_parent(parent, planetB):
        union(parent, planetA, planetB)
        result += cost
        
print(result)
```

```python
# 풀이 2. 시간 단축 방법

'''
1. 입력을 받은 뒤에 x축, y축, z축을 기준으로 각각 정렬을 수행
2. 각 축마다 n-1개의 간선만 확인해도 최적의 솔루션을 찾을 수 있음
따라서, 문제 풀이를 위해 고려하는 간선의 개수는 3*(n-1)개임 
'''

def find_parent(parent, x):
    if parent[x] != x:
        parent[x] = find_parent(parent, parent[x])
    return parent[x]

def union(parent, a, b):
    a = find_parent(parent, a)
    b = find_parent(parent, b)
    if a < b:
        parent[b] = a
    else:
        parent[a] = b

n = int(input())  # 행성의 개수

# 모든 노드에 대한 좌표값 입력받기
x = []
y = []
x = []
for i in range(1, n+1):
    data = list(map(int, input().split()))
    x.append((data[0],i))
    y.append((data[1],i))
    z.append((data[2],i))
x.sort()
y.sort()
z.sort()
    
parent = [0] * (n+1)
for i in range(n+1):
    parent[i] = i

edges = []
for i in range(n-1):
    edges.append((x[i+1][0]-x[i][0], x[i][1], x[i+1][1]))
    edges.append((y[i+1][0]-y[i][0], y[i][1], y[i+1][1]))
    edges.append((z[i+1][0]-z[i][0], z[i][1], z[i+1][1]))
edges.sort()    

result = 0
for edge in edges:
    cost, planetA, planetB = edge
    if find_parent(parent, planetA) != find_parent(parent, planetB):
        union(parent, planetA, planetB)
        result += cost
        
print(result)
```
